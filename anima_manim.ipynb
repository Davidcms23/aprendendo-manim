### 1.2 Primitivas geométricas

O Manim possui diversas formas básicas (Mobjetics) já prontas. Exemplo:

 - **Circle**: Cria um círculo.
 - **Square**: Cria um quadrado.
 - **Line**: Cria uma linha.

Para que as formas apareçam na tela, precisamos instruir a cena a exibi-las usando `self.add()` (para surgir instantaneamente) ou através de uma animação como `self.play(Create(forma))` (para desenhar a forma de maneira fl)

```python
%%manim -pql FormasBasicas

class FormasBasicas(Scene):
    def construct(self):
        circulo = Circle()
        quadrado = Square()
        linha = Line()

        self.play(Create(circulo))
        self.wait(1)

        self.play(Create(quadrado))
        self.play(Create(linha))
```

### 1.3 Posicionamento no plano

Se renderizar várias formas sem movê-las, todas aparecerão sobrepostas no centro da tela. Para organizar o espaço de trabalho, o Manim oferece métodos práticos para posicionar os objetos (Mobjetcs).:

 - `next_to`: Posiciona o objeto em relação a um ponto de referência ou outro objeto. O primeiro argumento é o objeto de referência e o segundo argumento é a direção, como `RIGHT` (direita), `LEFT` (esquerda), `UP` (cima) ou `DOWN` (baixo).
 - `shift`: Move o objeto a partir da sua posição atual (um deslocamento relativo).
 -  `move_to`: Move o objeto para uma coordenada absoluta exata no plano da câmera.

```python
%%manim -pql PosicionamentoPlano

class PosicionamentoPlano(Scene):
    def construct(self):
        circulo = Circle(color = RED)
        quadrado = Square(color = RED)
        linha = Line(color = RED)

        circulo.move_to(LEFT*3)

        quadrado.next_to(circulo, RIGHT, buff=1.0)

        linha.shift(DOWN*2)

        self.play(
            Create(circulo),
            Create(quadrado),
            Create(linha)
        )
```

```python
%%manim -pql PosicionamentoPlano2

class PosicionamentoPlano2(Scene):
    def construct(self):
        circulo = Circle(color = RED)
        quadrado = Square(color = BLUE)
        linha = Line(LEFT, RIGHT)

        circulo.move_to(LEFT*3)
    
        self.play(Create(circulo))
        self.play(Create(quadrado))

        linha.move_to(RIGHT*3)
        self.play(Create(linha))
```

## 2 Animações e tempo 

### 2.1 Adicionando Animações 

### 2.1.1 self.play() vs self.add() (Animar vs adicionar sileciosamente)

A diferença principal entre exibir algo na tela é se queremos que surja do nada ou seja desenhado:

 - `self.add(objeto)`: Adiciona o objetog à cena instanteneamete de forma silencioasa, sem nenhuma transição visual.
 - `self.play(Animacao(objeto))`: Executa uma animação na tela. Este método é a base do Manim e cria a transição visual ao longo do tempo.

#### 2.1.2 Animações de entrada e saída

Cada classe de animação encapsula um comportamento visual específico. Algumas das mais populares incluem:

 - `Create`: Desenha as bordas da forma.
 - `FadeIn`/`FadeOut`: Faz objeto aparecer ou desaparecer suavemente(dissolve) alterando sua opcacidade.
 - `GrowFromCenter`: O objeto expande a partir do seu ponto central até o tamanho normal.
 - `Write`: Simula o efeito de escrita à mão, sendo excelente para textos e fórmulas matemáticas.

### 2.2 Controle de Tempo

#### 2.2.1 run_time, self.wait() e rate_func

A velocidade com que as animações acontecem é vital para o ritmo do vídeo.

 - `self.wait(tempo)`: Pausa a cena por um determinado número de segundos antes de passar para a próxima linha de código.
 - `run_time`: É um parâmetro opcional(kwarg) passado dentro do `self.play()` que define a duração exata da animação em segundos. O padrão costuma ser 1 segundo.
 - `rate_func`: Define a "curva de velocidade" (easing) da animação. O padrão é `smooth`(começa devagar, acelara e termina devagar). Outros valores comuns incluem `linear`(velocidade constante) ou `there_and_back` (vai e volta).

### 2.3 Trnasformações entre objetos

Um dos maiores atrativos do Manim é a sua capacidade de transmutar formas geométricas suavemente usando poucas linhas de códigos. Ele possui duas funções principais para isso, e a diferença entre elas é sutil, mas muito importante.

#### 2.3.1 Transform

 - `Transform(mob1, mob2)`: Transforma os pontos e atributos (como cor) do `mob1` para que ele fique com a aparência do `mob2`. No entanto, internamente, o objeto em cena continua sendo a variável `mob1`.

#### 2.3.2 ReplacementTransform

 - `ReplacementTransform(mob1, mob2)`: Transforma `mob1` no `mob2` e **literalmente substitui** o `mob1` pelo `mob2` na memória da cena. Isso é muito útil se você for continuar aplicando animações ao `mob2`em passos subsequentes e não quiser lidar com referências antigas.

#### Exercício 2.3: Transformações Sucessivas

```python
# Responda aqui
```

```python
%%manim -pql Transformacoes

class Transformacoes(Scene):
    def construct(self):
        circulo = Circle(color=BLUE)
        quadrado = Square(color=BLUE)
        
        self.play(Create(circulo))
        self.play(ReplacementTransform(circulo, quadrado))
        self.play(quadrado.animate.set_color(YELLOW))
        
```

### 2.4 Animações simultânes e defasadas

Até agora, executamos animações de forma sequencial (uma após a outra, chamando `self.play()` várias vezes) ou todas exatamente ao mesmo tempo (passando várias animações seperadas por vírgula dentro de um único `self.play()`).

Para um controle melhor de agrupamento, usamos `AnimationGroup`e `LaggedStart`.

#### 2.4.1 AnimationGroup

 - `AnimationGroup`: Agrupa várias animações para que o Manim as execute juntas como um bloco único simultâneo.

#### 2.4.2 LaggedStart

 - `LaggedStart`:É uma variação do `AnimationGroup`. Ele inicia as animações sequencialmente, mas com uma sobreposição (defasagem) entre elas, gerando um efeito de cascata visualmente muito agradável. Você controla esse atraso usando o argumento `lag_ratio`.

#### Exemplo

```python
%%manim -pql Agrupamento

class Agrupamento(Scene):
    def construct(self):
        circulo1 = Circle(color=RED).shift(LEFT*2, UP*1.5)
        quadrado1 = Square(color=RED).shift(RIGHT*2, UP*1.5)

        circulo2 = Circle(color=GREEN).shift(LEFT*2, DOWN*1.5)
        quadrado2 = Square(color=GREEN).shift(RIGHT*2, DOWN*1.5)

        grupo = AnimationGroup(
            Create(circulo1),
            Create(quadrado1)
        )
        self.play(grupo)
        self.wait(1)
        
        cascata = LaggedStart(
            Create(circulo2),
            Create(quadrado2)
        )
        self.play(cascata)
```

## 3 Cores, Estilos e Grupos

### 3.1 Preenchimento e contorno

#### 3.1.1 set_color() / set.fill() / set_stroke()

Por padrão, muitas formas geométricas no Manim possuem apenas um contorno. Podemos alterar suas propriedades visuais a qualquer momento usando métodos específicos:

- `set_color(cor)`: Altera a cor geral do objeto (útil para linhas e textos).
- `set_fill(cor, opacity)`: Define a cor de preenchimento interno do objeto e sua opacidade (de 0.0 a 1.0).
- `set_stroke(cor, width)`: Define a cor e a espessura da linha de contorno (a borda) da forma.

#### Exemplo

```python
%%manim -pql ExemploEstilos

class ExemploEstilos(Scene):
    def construct(self):
        # Cria um círculo e aplica os estilos
        circulo = Circle(radius=2)
        
        # Modificando o estilo diretamente
        circulo.set_stroke(color=BLUE, width=10)
        circulo.set_fill(color=PINK, opacity=0.8)
        
        self.play(Create(circulo))
        self.wait(1)
```
