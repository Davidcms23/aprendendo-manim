### 1.2 Primitivas geométricas

O Manim possui diversas formas básicas (Mobjetics) já prontas. Exemplo:

 - **Circle**: Cria um círculo.
 - **Square**: Cria um quadrado.
 - **Line**: Cria uma linha.

Para que as formas apareçam na tela, precisamos instruir a cena a exibi-las usando `self.add()` (para surgir instantaneamente) ou através de uma animação como `self.play(Create(forma))` (para desenhar a forma de maneira fl)

```python
%%manim -pql FormasBasicas

class FormasBasicas(Scene):
    def construct(self):
        circulo = Circle()
        quadrado = Square()
        linha = Line()

        self.play(Create(circulo))
        self.wait(1)

        self.play(Create(quadrado))
        self.play(Create(linha))
```

### 1.3 Posicionamento no plano

Se renderizar várias formas sem movê-las, todas aparecerão sobrepostas no centro da tela. Para organizar o espaço de trabalho, o Manim oferece métodos práticos para posicionar os objetos (Mobjetcs).:

 - `next_to`: Posiciona o objeto em relação a um ponto de referência ou outro objeto. O primeiro argumento é o objeto de referência e o segundo argumento é a direção, como `RIGHT` (direita), `LEFT` (esquerda), `UP` (cima) ou `DOWN` (baixo).
 - `shift`: Move o objeto a partir da sua posição atual (um deslocamento relativo).
 -  `move_to`: Move o objeto para uma coordenada absoluta exata no plano da câmera.

```python
%%manim -pql PosicionamentoPlano

class PosicionamentoPlano(Scene):
    def construct(self):
        circulo = Circle(color = RED)
        quadrado = Square(color = RED)
        linha = Line(color = RED)

        circulo.move_to(LEFT*3)

        quadrado.next_to(circulo, RIGHT, buff=1.0)

        linha.shift(DOWN*2)

        self.play(
            Create(circulo),
            Create(quadrado),
            Create(linha)
        )
```

```python
%%manim -pql PosicionamentoPlano2

class PosicionamentoPlano2(Scene):
    def construct(self):
        circulo = Circle(color = RED)
        quadrado = Square(color = BLUE)
        linha = Line(LEFT, RIGHT)

        circulo.move_to(LEFT*3)
    
        self.play(Create(circulo))
        self.play(Create(quadrado))

        linha.move_to(RIGHT*3)
        self.play(Create(linha))
```

## 2 Animações e tempo 

### 2.1 Adicionando Animações 

### 2.1.1 self.play() vs self.add() (Animar vs adicionar sileciosamente)

A diferença principal entre exibir algo na tela é se queremos que surja do nada ou seja desenhado:

 - `self.add(objeto)`: Adiciona o objetog à cena instanteneamete de forma silencioasa, sem nenhuma transição visual.
 - `self.play(Animacao(objeto))`: Executa uma animação na tela. Este método é a base do Manim e cria a transição visual ao longo do tempo.

#### 2.1.2 Animações de entrada e saída

Cada classe de animação encapsula um comportamento visual específico. Algumas das mais populares incluem:

 - `Create`: Desenha as bordas da forma.
 - `FadeIn`/`FadeOut`: Faz objeto aparecer ou desaparecer suavemente(dissolve) alterando sua opcacidade.
 - `GrowFromCenter`: O objeto expande a partir do seu ponto central até o tamanho normal.
 - `Write`: Simula o efeito de escrita à mão, sendo excelente para textos e fórmulas matemáticas.

### 2.2 Controle de Tempo

#### 2.2.1 run_time, self.wait() e rate_func

A velocidade com que as animações acontecem é vital para o ritmo do vídeo.

 - `self.wait(tempo)`: Pausa a cena por um determinado número de segundos antes de passar para a próxima linha de código.
 - `run_time`: É um parâmetro opcional(kwarg) passado dentro do `self.play()` que define a duração exata da animação em segundos. O padrão costuma ser 1 segundo.
 - `rate_func`: Define a "curva de velocidade" (easing) da animação. O padrão é `smooth`(começa devagar, acelara e termina devagar). Outros valores comuns incluem `linear`(velocidade constante) ou `there_and_back` (vai e volta).

### 2.3 Transformações entre objetos

Um dos maiores atrativos do Manim é a sua capacidade de transmutar formas geométricas suavemente usando poucas linhas de códigos. Ele possui duas funções principais para isso, e a diferença entre elas é sutil, mas muito importante.

#### 2.3.1 Transform

 - `Transform(mob1, mob2)`: Transforma os pontos e atributos (como cor) do `mob1` para que ele fique com a aparência do `mob2`. No entanto, internamente, o objeto em cena continua sendo a variável `mob1`.

#### 2.3.2 ReplacementTransform

 - `ReplacementTransform(mob1, mob2)`: Transforma `mob1` no `mob2` e **literalmente substitui** o `mob1` pelo `mob2` na memória da cena. Isso é muito útil se você for continuar aplicando animações ao `mob2`em passos subsequentes e não quiser lidar com referências antigas.

### 2.4 Animações simultânes e defasadas

Até agora, executamos animações de forma sequencial (uma após a outra, chamando `self.play()` várias vezes) ou todas exatamente ao mesmo tempo (passando várias animações seperadas por vírgula dentro de um único `self.play()`).

Para um controle melhor de agrupamento, usamos `AnimationGroup`e `LaggedStart`.

#### 2.4.1 AnimationGroup

 - `AnimationGroup`: Agrupa várias animações para que o Manim as execute juntas como um bloco único simultâneo.

#### 2.4.2 LaggedStart

 - `LaggedStart`:É uma variação do `AnimationGroup`. Ele inicia as animações sequencialmente, mas com uma sobreposição (defasagem) entre elas, gerando um efeito de cascata visualmente muito agradável. Você controla esse atraso usando o argumento `lag_ratio`.

#### Exemplo

```python
%%manim -pql Agrupamento

class Agrupamento(Scene):
    def construct(self):
        circulo1 = Circle(color=RED).shift(LEFT*2, UP*1.5)
        quadrado1 = Square(color=RED).shift(RIGHT*2, UP*1.5)

        circulo2 = Circle(color=GREEN).shift(LEFT*2, DOWN*1.5)
        quadrado2 = Square(color=GREEN).shift(RIGHT*2, DOWN*1.5)

        grupo = AnimationGroup(
            Create(circulo1),
            Create(quadrado1)
        )
        self.play(grupo)
        self.wait(1)
        
        cascata = LaggedStart(
            Create(circulo2),
            Create(quadrado2)
        )
        self.play(cascata)
```

### 3.2 Agrupamento de objetos

#### 3.2.1 VGroup e arrange()

Muitas vezes, queremos organizar várias formas juntas para movê-las ou animá-las de uma só vez.

- `VGroup`: É uma classe que permite agrupar múltiplos `Mobjects` (Objetos do Manim) em uma única coleção.
- `arrange(direction, buff)`: Um método poderoso que, quando aplicado a um `VGroup`, alinha automaticamente todos os itens do grupo em uma direção específica (como `RIGHT` ou `DOWN`), com um espçamento (`buff`) entre eles.

Exemplo:

```python
%%manim -pql ExemploGrupo

class ExemploGrupo(Scene):
    def construct(self):
        c1 = Circle(radius=.5, color=RED).set_fill(RED, 1)
        c2 = Circle(radius=.5, color=GREEN).set_fill(GREEN, 1)
        c3 = Circle(radius=.5, color=BLUE).set_fill(BLUE, 1)

        grupo = VGroup(c2, c1, c3) # verde, vermelho e azul

        grupo.arrange(RIGHT, buff=.5)

        self.play(DrawBorderThenFill(grupo))
        self.wait()
```

### 3.3 Animar propriedades e destaque

#### 3.3.1 A sintaxe .animate

O `.animate` é um dos recursos mais poderosos do Manim. Quando você adiciona `.animate`antes de qualquer método que modifique um objeto (como `shift`, `scale`, `set_fill`, etc.), o Manim transforma essa modificação direta em uma animação fluida, interpolando o estado inicial e o estado final do objeto. Ele deve ser executado sempre dentro de um `self.play()`.

#### 3.3.2 Indicate e Flash

Além de modificar propriedades, às vezes precisamos chamar a atenção do espectador para um objeto específico na tela usando animações de indicação:

 - `Indicate(objeto)`: Destaca o objeto, mudando temporariamnete sua cor e aumentando sua escala, chamando atenção para ele.
 - `Flash(objeto)`: Cria pequenas linhas animadas (faíscas) que irradiam a partir do centro do objeto.

Exemplo

```python
%%manim -pql ExemploDestaque

class ExemploDestaque(Scene):
    def construct(self):
        quadrado = Square().set_fill(BLUE, 1)
        self.add(quadrado)
        self.wait(.5)

        self.play(quadrado.animate.set_fill(RED).shift(UP*2))
        self.play(quadrado.animate.set_fill(GREEN).shift(DOWN*2))
        self.play(quadrado.animate.set_fill(RED).shift(UP*2))
        
        self.play(Indicate(quadrado))
        self.play(Flash(quadrado))
```

## 4 Texto e $\text{LaTeX}$

O Manim foi originalmente projetado para criar videos matemáticos explicativos, o que significa que ele possui um suporte excelente para renderizar textos e equações. Se você precisar de fórmulas matemáticas avançadas na sua animação, precisará ter o pacote LaTeX configurado no seu sistema. Neste módulo, aprenderemos como escrever textos, renderizar fórmulas bonitas e, o mais impressionante, como animar a transição lógica entre diferentes equações.

### 4.1 Três classes de texto

#### 4.1.1 Text/Tex/MathTex

No Manim, a renderização de palavras e fórmulas é feita principalmente através de três classes distintas:

 - `Text`: Usado para texto comum. Não precisa compilador LaTex e permite o uso fácil de fontes normais do seu sistema.
 - `Tex`: Usado para renderizar texto utilizando o LaTeX. É excelente quando você quer o visual clássico e acadêmico do LaTeX em parágrafo normais.
 - `MathTex`: Usado especificamente para renderizar fórmulas matemáticas. Ele funciona como se todo o texto estivesse automaticamente dentro do ambiente de equações do LaTeX (ou seja, não é necessário colocar sinais de cifrão `$` no início e no final).

**Nota importante**: Ao usar `Tex` ou `MathTex`, é uma boa prática usar raw strings no python (colocando um `r` antes das aspas, ex: `r"x^2") para evitar que barras invertidas sejam confundidas com caracteres de escapa.

Exemplo:

```python
%%manim -pql ExemploClassesDeTexto

from manim import *

class ExemploClassesDeTexto(Scene):
    def construct(self):
        texto_comum = Text("Teorema de Pitágoras:", font_size=36)
        formula = MathTex(r"a^2 + b^2 = c^2", font_size=48)
        texto_latex = Tex(r"Onde $c$ é a hipotenusa.", font_size=32)

        texto_comum.move_to(UP*2)
        formula.next_to(texto_comum, DOWN, buff=.5)
        texto_latex.next_to(formula, DOWN, buff=.5)

        self.play(Write(texto_comum))
        self.play(FadeIn(formula, shift=UP))
        self.play(Write(texto_latex))
        self.wait(2)
```

### 4.2 Colorir partes da equação

#### 4.2.1 MathTex[] e set_color_by_tex

Fórmulas monocromátricas podem ser difíceis de ler. Uma técnica muito comum no Manim é destacar variáveis específicar usando cores para facilitar a explicação.

Para podermos colorir partes separadas de uma fórmula, não podemos passar a equação inteira como uma única string para o `MathTex`. Em vez disso, quebramos a equação em uma lista de strings. O `MathTex` vai juntar tudo na hora de desenhar, mas tratará cada parte como um pomponente independente (acessível via indice, como listas normais no Python) ou referenciável pelo seu conteúdo.

O método `set_color_by_tex("texto", COR)` varre as strings que compõem o seu `MathTex` e colore qualquer bloco que corresponda à string passada.

Exemplo:

```python
%%manim -pql ExemploColorirEquacao


class ExemploColorirEquacao(Scene):
    def construct(self):
        equacao = MathTex("f(x)", "=", "x^2", "+", "3x")
        
        equacao.set_color_by_tex("f(x)", BLUE)
        equacao.set_color_by_tex("x^2", GREEN)
        equacao[3].set_color(RED)
        equacao.set_color_by_tex("3x", YELLOW)
        
        self.play(Write(equacao))
        self.play(equacao.animate.scale(1.5))
        self.wait(2)
```

### 4.3 Transformar equações com correspondência

#### 4.3.1 TransformMatchingTex

No módulo 2 vimos as transições `Transform` e `ReplacementTransform`. Elas funcionam bem para formas geométricas inteiras, mas ao transformar equações longas, os caracteres podem "voar" pela tela de maneira caótica.

Para resolver a matemática passo a passo e de forma fluida, o Manim oferece a classe de animação abstrata `TransfromMatchingTex`. Se você tiver duas fórmulas quebradas em substrings (como ensinado no passo anterior), essa animação irá identificar as strings em comum entre a primeira equação e a segunda, movendo-as suavemente e preservando suas cores e formatos.

Exemplo:

```python
%%manim -pql ExemploTransformacaoEquacao

class ExemploTransformacaoEquacao(Scene):
    def construct(self):
        eq1 = MathTex("a", "+", "b", "=", "c")
        eq2 = MathTex("a", "=", "c", "-", "b")
        
        eq1.set_color_by_tex("b", YELLOW)
        eq2.set_color_by_tex("b", YELLOW)

        self.play(Write(eq1))
        self.wait(1)
        
        self.play(TransformMatchingTex(eq1, eq2))
        self.wait(2)
```

## 5 Eixos, Gráficos e Funções

### 5.1 Criar o plano cartesiano

#### 5.1.1 Axes vs NumberPlane

Dentro do módulo de sistemas de coordenadas (`coordinate_systems`), existem duas formas principais de exibir um plano 2D na tela:

 - `Axes`: Cria dois eixos simples (X e Y), sem linhas de grade. Você pode configurar o alcance de cada eixo (valores mínimos e máximos) e a distância entre os mercadores.
 - `NumberPlane`: Cria um plano cartesiano completo, com linhas de grade ao fundo e eixos centrais. É excelente para dar uma noção espacial e escalar ao espectador.

Exemplo:

```python
%%manim -pql ExemploEixos

class ExemploEixos(Scene):
    def construct(self):
        eixos = Axes(
            x_range=[-3, 3, 1], # Altere os valores e observe o mudança
            y_range=[-2, 2, 1],
            x_length=6, 
            y_length=6
        )
        
        rotulos = eixos.get_axis_labels(x_label="x", y_label="y")
        
        self.play(Create(eixos), Write(rotulos))
        self.wait(2)
```

### 5.3 Preencher área sob a curva

#### 5.3.1 ax.get_area()

Para videos de cálculo abordando integrais, destacar a área delimitada entre a curva da função e o eixo X é fundamental. Usamos o método `ax.get_area(grafico, x_range=[início, fim]`) para gerar um objeto (uma forma preenchida) que representa perfeitamente essa região. Você também pode difinir a cor e a opacidade dessa área.

```python
%%manim -pql ExemploArea

class ExemploArea(Scene):
    def construct(self):
        ax = Axes(
            x_range=[-10, 10, 1],
            y_range=[0, 50, 5]
        )
        grafico = ax.plot(lambda x: .5*x**2, color=GREEN)

        area = ax.get_area(grafico, x_range=[0, 10], color=BLUE, opacity=.5)

        self.add(ax)
        self.wait()
        self.play(Create(grafico))
        self.wait()
        self.play(FadeIn(area))
        self.wait()
```

## 6 ValueTrackers e updaters

### 6.1 Variável animável

#### 6.1.1 ValueTracker

O `ValueTracker` é um Mobject invisível que guarda um valor numérico. A mágica acontece porque podemos animar a mudança desse valor usando o método `.animate.set_value(novo_valor)`. Quando combinamos um `ValueTracker` com "updaters" (que veremos a seguir), conseguimos atualizar a tela inteira com base nesse único número, garantindo uma transição fluida.

```python
%%manim -pql ExemploTracker

class ExemploTracker(Scene):
    def construct(self):
        rastreador = ValueTracker(0)
        numero_tela = always_redraw(lambda: DecimalNumber(rastreador.get_value(), num_decimal_places=1).scale(2))

        self.add(numero_tela)

        self.play(rastreador.animate.set_value(100), run_time=3)
        self.wait()
```

### 6.2 Vincular objeto a uma função de atualização

#### 6.2.1 add_update()

O método `.add_updaters()` permite que você anexe uma função a uma Mobject. Essa função será chama automaticamente pelo Manim em cada frame da animação (geralmente 60 vezes por segundo). Isso é perfeito para fazer com que um objeto persiga outro ou reaja a mudança na cena de maneira condicional.

A função que você passa para o `add_updater` sempre recebe o próprio Mobject como primeiro argumento (costumamos chamá-lo de `m` ou `mob`).

```python
%%manim -pql ExemploUpdater

class ExemploUpdater(Scene):
    def construct(self):
        ponto = Dot(LEFT*4, color=BLUE).scale(2)
        texto = Text("Me siga!").scale(.6)

        self.add(texto)
        self.wait()
        self.play(FadeIn(ponto))
        self.wait()
        
        self.play(texto.animate.next_to(ponto, UP))
        texto.add_updater(lambda m: m.next_to(ponto, UP))
        
        self.play(ponto.animate.shift(RIGHT*10), run_time=2)
        self.wait()
```
